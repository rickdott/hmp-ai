```python
import hsmm_mvpy as hmp
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from hmpai.visualization import set_seaborn_style
```

```python
fit = hmp.utils.load_fit("../models/hmp/500hz_accuracy.nc")
```

```python
set_seaborn_style()
# colors = sns.color_palette("tab10", n_colors=len(fit.eventprobs.event))


for event in fit.eventprobs.event:
    event_data = fit.eventprobs.isel(trial_x_participant=25, event=event)
    plt.axvline(
        x=event_data.idxmax().item(),
        ymin=0.05,
        ymax=0.99,
        color=sns.color_palette()[event.item()],
        linestyle="--",
    )
    plt.plot(event_data, color=sns.color_palette()[event.item()])

plt.xlabel("Time (in samples)")
plt.ylabel("Probability of transition")
sns.despine()
# add vertical line at maximum Y value
# max_y = max([max(event_data) for event_data in fit.eventprobs.values()])
# plt.axvline(x=max_y, color='r')

plt.show()
```
